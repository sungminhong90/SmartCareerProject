```
import matplotlib.pyplot as plt    
%matplotlib inline 
from wordcloud import WordCloud, STOPWORDS



```

```
import numpy as np
from PIL import Image
```

```
text = open("all_skills.txt").read()
data_mask = np.array(Image.open("dataanalysis.PNG"))
stopwords = set(STOPWORDS)
stopwords.add(",")
```

```
!ls
```

```
wc = WordCloud(background_color='white',
               max_words = 200,
               mask= data_mask,
               contour_width=0.2,
               contour_color='green',
               max_font_size=700,
               random_state=1)
wc.generate(text)
wc.words_

```
