# skill table 분리

```python
df_se = df[['Skills & Endorsements']]
df_tt = df[['Tools & Technologies']]
df_is = df[['Interpersonal Skills']]
df_ik = df[['Industry Knowledge']]
df_os = df[['Other Skills']]
```

# 데이터 reshape

```python
df_se = df_se['Skills & Endorsements'].apply(lambda x: pd.Series(x))
df_tt = df_tt['Tools & Technologies'].apply(lambda x: pd.Series(x))
df_is = df_is['Interpersonal Skills'].apply(lambda x: pd.Series(x))
df_ik = df_ik['Industry Knowledge'].apply(lambda x: pd.Series(x))
df_os = df_os['Other Skills'].apply(lambda x: pd.Series(x))
```

```python
df_se.head()
```

```python
df_tt.head()
```

```python
df_se = df_se.stack().reset_index(level=1, drop=True).to_frame('Skills & Endorsements')
df_tt = df_tt.stack().reset_index(level=1, drop=True).to_frame('Tools & Technologies')
df_is = df_is.stack().reset_index(level=1, drop=True).to_frame('Interpersonal Skills')
df_ik = df_ik.stack().reset_index(level=1, drop=True).to_frame('Industry Knowledge')
df_os = df_os.stack().reset_index(level=1, drop=True).to_frame('Other Skills')
```

```python
df_se.head()
```

```python
df_tt.head()
```

```python
df_se['ID'] = df_se.index
df_tt['ID'] = df_tt.index
df_is['ID'] = df_is.index
df_ik['ID'] = df_ik.index
df_os['ID'] = df_os.index
```

```python
df_se.head()
```

# 각 skill 데이터 저장 (csv)

```python
df_se.to_csv('Skills & Endorsements.csv', index=False)
df_tt.to_csv('Tools & Technologies.csv', index=False)
df_is.to_csv('Interpersonal Skills.csv', index=False)
df_ik.to_csv('Industry Knowledge.csv', index=False)
df_os.to_csv('Other Skills.csv', index=False)
```
